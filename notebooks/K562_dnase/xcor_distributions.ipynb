```python
fg_path = "/oak/stanford/groups/akundaje/atwang/finemo_gpu_test/K562/fg_calibration/distributions.npz"
bg_path = "/oak/stanford/groups/akundaje/atwang/finemo_gpu_test/K562/background/calibration/distributions.npz"

fg_data = np.load(fg_path)
bg_data = np.load(bg_path)
```

```python
def plot_xcor_hist(fg_data, bg_data, motif_name):

    fg_xcors = fg_data[motif_name]
    bg_xcors = bg_data[motif_name]

    plt.hist(fg_xcors, density=True, bins="auto", range=(0, 2.5), alpha=0.4, label="Peaks",)
    plt.hist(bg_xcors, density=True, bins="auto", range=(0, 2.5), alpha=0.4, label="Background")

    plt.title(f"Distribution of maximum {motif_name} cross-correlation per region")
    plt.xlabel("Arctanh cross-correlation")
    plt.ylabel("Frequency")

    plt.legend()

    plt.show()
    plt.close()
```
